# Extraer Datos de la Web (sin logging)

```python
#Importar librer√≠as
import requests
```

```python
#Definir parametros 
url = 'https://www.bcentral.cl/inicio'
page = requests.get(url)
page
#Response 200: se hizo solicitud a pagina de forma exitosa
```

```python
#Imprimir string de la pagina web
page.content
```

```python
#Importar librer√≠a para ver contenido web
from lxml import html
```

```python
#Variable que contenga contenido de web
tree = html.fromstring(page.content)
tree
```

```python
#Metodo Xpath para acceder a la informaci√≥n que nos importa
tree.xpath('//*[@id="_BcentralIndicadoresViewer_INSTANCE_pLcePZ0Eybi8_myTooltipDelegate"]/div/div/div[1]/div/div/div[3]/div/p[2]')
```

```python
#Agregar  metodo /text() al final de la cadena Xpath , porque requiero obtener solo el texto 
tree.xpath('//*[@id="_BcentralIndicadoresViewer_INSTANCE_pLcePZ0Eybi8_myTooltipDelegate"]/div/div/div[1]/div/div/div[3]/div/p[2]/text()')
```

```python
#Acceder al primer elemento
usd = tree.xpath('//*[@id="_BcentralIndicadoresViewer_INSTANCE_pLcePZ0Eybi8_myTooltipDelegate"]/div/div/div[1]/div/div/div[3]/div/p[2]/text()')
usd[0]
```

```python
#Separar por caracter(espacio)
usd[0].split(' ')
```

```python
#Seleccionar primer elemento
usd[0].split(' ')[0]
```

```python
#Retirar simbolo de $
usd[0].split(' ')[0].strip("$")
```

```python
#Cambiar , por punto
usd[0].split(' ')[0].strip("$").replace(",",".")
```

```python
#Convertir a Numero para su posterior manipulaci√≥n
float(usd[0].split(' ')[0].strip("$").replace(",","."))
```

```python
#Crear variable USD
usd = float(usd[0].split(' ')[0].strip("$").replace(",","."))
```

## Actualizar data en archivo local

```python
#Abrir archivo excel
wb = load_workbook('calculadoratoy.xlsx')
```
